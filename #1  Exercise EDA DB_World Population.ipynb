## All variables correlation by chart (Heatmap)

```python
plt.figure(figsize=(10, 9))  
correlation_matrix = df.corr()  

sns.heatmap(correlation_matrix,
            annot=True,          
            cmap='vlag',     
            center=0,            
            #vmin=-1,             
            #vmax=1,              
            square=True,         
            linewidths=0, 
            linecolor= 'gray',
            cbar_kws={"shrink": .5})

plt.title('Correlation Heatmap', fontsize=9)

plt.tight_layout()  # Adjust layout

plt.show()
```

## For correlation between 2 variables

```python
g = sns.relplot(data=df, x="Density (per km²)", y="2022 Population")

g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(8, 4))
```

**| Look for results by continent group by 'mean'.Then is sorted out as a descending order**

```python
df.groupby(df['Continent']).mean().sort_values(by = '2022 Population',ascending = False)
```

**|Show me only the top 5 'Oceania' continent results**

```python
df.columns
```

```python
df.groupby('Country').max().head(10).sort_values(by = '2022 Population', ascending = False)
```

 **| Access only the columns that you need, in this case from 6 to 13**

```python
df.groupby(df['Continent'])[df.columns[5:13]].mean().sort_values(by = '2022 Population',ascending = False)
```

**|How to aggregate information. Here I created 3 new columns by the variable I wanted to compare, maintaining the group by country**

```python
df.groupby('Country').agg({'2022 Population':['mean','max','min'],'2020 Population':['mean','max','min'],'Density (per km²)':['mean','max','min'], 'Growth Rate':['mean','max','min']})
```
