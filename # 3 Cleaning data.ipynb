# 5. Start Cleaning

## 5.1 Drop off duplicates _ ONLY if needed

```python
# df.drop_duplicates() eliminate those duplicates
# df = df.drop_duplicates() saves those changes !!

df = df.drop_duplicates()
df
```

## 5.2 Drop off NOT neccesary columns

```python
# df.drop (columns = 'Not_Useful_Column') to remove unwanted columns_ I already saved it
df
```

## 5.3 Using 'strip'to remove spaces or unwanted characters

```python
# Cleaning the 'last_Name' column
# String takes off spaces , right and left

df['Last_Name']= df['Last_Name'].str.lstrip ('...')
df['Last_Name']= df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.rstrip('_')
df
```

## 5.4 Using 'replace' to clean unwanted characters among others

```python
#Leave only numbers for the column 'Phone_Number'

#df['Phone_Number'].str.replace ('[^a-zA-Z0-9]','')

# I runned it and it worked so now we can save it

#df['Phone_Number'] = df['Phone_Number'].str.replace ('[^a-zA-Z0-9]','')
```

```python
# Dropped it cause its no longer I column to work on.

df= df.drop (columns ='Phone_Number')
df
```

## 5.5 Splitting columns

The adress can be separated in 3 columns [Number_avenue_postcode] an they have comma

```python
df[['Street_Address','Street','Zip_Code']]=df['Address'].str.split(',',2,expand = True)
```

```python
df
```

## 5.6 Standarize [Yes/Y | No/N]

```python
df['Paying Customer']= df['Paying Customer'].str.replace ('Yes','Y')
df['Paying Customer']= df['Paying Customer'].str.replace ('No','N')
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df.isnull().sum()
```

## 5.7 Replace for the ENTIRE dataframe (df)

```python
df = df.replace('N/a','')
```

```python
# NaN is not a value , technically is empty

# fillna fill every space with a blank | I used the 'inplace' to remember the use

df.fillna('',inplace = True)
```

```python
df
```

```python
# Cleaning characters around the 'last names'

df['Last_name']= df['Last_name'].str.lstrip('/')
df['Last_name']= df['Last_name'].str.rstrip('_')
df
```

```python
# Dropping the adress cause we alreay have new 3 columns

df.drop (columns = 'Address',inplace = True)
```

## 5.8 Leave only numbers allow to call by 'Y'

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop (x,inplace=True)
    
df

#This would be the list of contacts to call, 
#they either directly responden NO when asked 'not called' or did not respond.
```
